## Exercise 1:  Define an Asteroid class

Define a simple Asteroid class with some basic state variables for a single NEO.  Your Asteroid class should define at least 4 "state variables:”

    id
    name
    estimated_diameter (m)
    is_potentially_hazardous (bool)

Operations must include:
* `__init__(self, ...)` method to initialize a new Asteroid object with specific data values
* `__str__(self)`, and `__repr__(self)` methods that return nicely formatted string representations of the object.

Write a little code to test your new class.

````python
class Asteriod:

  def __init__(self, name, id:int, estimated_diameter:float, is_potentially_hazardous:bool):

    self.name = name
    self.id = id
    self.estimated_diameter = estimated_diameter
    self.is_potentially_hazardous = is_potentially_hazardous

  def getName(self):
    return self.name

  def getId(self):
    return self.id

  def getEstimated_Diameter(self):
    return self.estimated_diameter

  def getIs_Potentially_Hazardous(self):
    return self.is_potentially_hazardous

  def __str__(self):
    #the one made to be read by people:
    return (f'The Name of the asteriod is:{self.getName()} , the Id is {self.getId()} , an estimate of the diamter is:{self.getEstimated_Diameter()} , and it is {self.getIs_Potentially_Hazardous()} that it is potentialy hazardous')

  def __repr__(self):
    #a list made for using in other funcs
    return (f'Name:{self.getName(self)},ID:{self.getId(self)},Estimated_Diameter:{self.getEstimated_Diameter(self)},Is_potentially_Hazardous{self.getIs_Potentially_Hazardous(self)}')

#below is a generated asteriod imma use for testing
#test_asteroid = Asteroid(name="Sample Asteroid", id=12345, estimated_diameter=100.5, is_potentially_hazardous=False)
````

## Exercise 2: Make a histogram

Create a list of all the Asteroid objects from this week's NEO feed and use the list to create a histogram to visualize the distribution of Asteroid diameters.  To create the historgram use the `hist()` method in the `matplotlib.pyplot` module.

````python
# Ex. 2 your code here

from datetime import timedelta

def create_asteroid_objects(neos_data):

    asteroids = []

    for neo in neos_data:
        diameter = neo['estimated_diameter']['meters']['estimated_diameter_max']
        asteroid = Asteroid(
            name=neo['name'],
            id=neo['id'],
            estimated_diameter=diameter,
            is_potentially_hazardous=neo['is_potentially_hazardous_asteroid']
        ) #gemini reformatted this part to look better since i asked it to correct a mistake, i missed a comma inside it before corrections
        asteroids.append(asteroid)
    return asteroids

def create_histogram(asteroids):
    diameters = [asteroid.estimated_diameter for asteroid in asteroids]
    plt.hist(diameters, bins=50)
    plt.xlabel('Estimated Diameter (meters)')
    plt.ylabel('Frequency')
    plt.title('Distribution of Asteroid Diameters')
    plt.show()

today = datetime.date.today()
start_date = today - datetime.timedelta(days=7)
neos_data = get_neos(start_date)
asteroids = create_asteroid_objects(neos_data)
create_histogram(asteroids)
````

 2. Use the `@classmethod` decorator to construct a `from_NEO()` 'factory method' inside your Asteroid class that creates Asteroid objects from the NEO API.  The method should take the id for a single NEO, fetche the NEO record from API, constructs a single Asteroid object, and return the Asteroid object representing that NEO.  This kind of method is called a “Factory” because it constructs an object from raw materials.  The object factory method should start like this:
    ```
    @classmethod
    from_NEO(cls, neo_id):
        ...
    ```
Write a little code to test your new method.

 3. Use your new Asteroid class to create a list of Asteroid objects.  Use this list to create the same historgram as in Exercise 2.  




````python
# Ex.3 your code here
import datetime

@dataclass
class Asteroid:
  name: str
  id: int
  estimated_diameter: float
  is_potentially_hazardous: bool

  @classmethod
  def from_NEO(cls, neo_id):

    neo = get_neo(neo_id)
    diameter = neo['estimated_diameter']['meters']['estimated_diameter_max']
    return cls(
        name=neo['name'],
        id=neo['id'],
        estimated_diameter=diameter,
        is_potentially_hazardous=neo['is_potentially_hazardous_asteroid']
    )
#everything above this comment is generated,  i need a better understanding of @dataclass and redefining classes
def create_histogram(asteroids):

    diameters = [asteroid.estimated_diameter for asteroid in asteroids]
    plt.hist(diameters, bins=20)
    plt.xlabel('Estimated Diameter (meters)')
    plt.ylabel('Frequency')
    plt.title('Distribution of Asteroid Diameters')
    plt.show()


today = datetime.date.today()
start_date = today - datetime.timedelta(days=7)
neos_data = get_neos(start_date)


asteroids = [Asteroid.from_NEO(neo['id']) for neo in neos_data]
create_histogram(asteroids)
````
